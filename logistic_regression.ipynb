```python
import pandas as pd
from sklearn.preprocessing import LabelBinarizer
```

```python
df = pd.read_csv("data/train.csv").drop(labels=["Name", "Ticket", "Cabin"], axis=1).dropna()
```

```python
df = pd.get_dummies(df)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df.drop(labels='Survived', axis=1),
                                                    df['Survived'],
                                                    test_size=0.2,
                                                    random_state=42)
```

```python
from lib.models import LogisticRegression


regr = LogisticRegression()
regr.train(x_train, y_train, max_epochs=50)
```

```python
from sklearn.linear_model import LogisticRegression as lg
from sklearn.metrics import accuracy_score


sklearn_regr = lg().fit(x_train, y_train)
print(f"Sklearn score is {accuracy_score(sklearn_regr.predict(x_test), y_test)}")
print(f"My score is {accuracy_score(regr.predict(x_test), y_test)}")
```
